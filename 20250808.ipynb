```python
def printfile(name):
    with open(name, 'r') as f:
      for line in f:
          print(line)
```

```python
printfile('sample.txt')
```

```python
def new_line(name, ln):
    with open(name, 'w') as f:
      for line in ln:
        f.write(line + '\n')
```

```python
fruits = ['grape', 'orange']
```

```python
new_line('output.txt', fruits)
```

```python
def arange_square_matrix(n):
    rows = [np.arange(i, i + n) for i in range(n)]
    return np.array(rows)
```
